## What to build
`inception(x, f1, f3r, f3, f5r, f5, fp, name)` in `cnn/model.py`. Four parallel branches on the
input tensor `x`, **concatenated on the channel axis**:
1. `1×1` conv, `f1` filters.
2. `1×1` (`f3r`) → `3×3` (`f3`).
3. `1×1` (`f5r`) → `5×5` (`f5`).
4. `3×3` MaxPool (stride 1, same) → `1×1` (`fp`).
All convs `padding="same"`, `relu`. Name layers `f"{name}_..."`. **Output channels = f1 + f3 + f5 + fp**,
spatial size unchanged. (The `1×1` reducers `f3r/f5r` are bottlenecks: shrink channels before the
costly 3×3 / 5×5.)
